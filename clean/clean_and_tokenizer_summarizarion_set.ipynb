```python
import pandas as pd
import re
import string
```

```python
df = pd.read_csv('/kaggle/input/nepali-summarization-set-raw/summarization_set_raw.csv')
original_dataset = len(df)
```

```python
### NA and Duplicates
df = df.dropna(subset=['news'])
df = df.dropna(subset=['title'])
df = df.drop_duplicates(subset=['title', 'news'])
```

```python
def perform_preprocessing(text):
    text = (
        text.replace("\n", " ")
        .replace("\r", " ")
        .replace("\t", " ")
        .replace("–", " ")
        .replace("-", " ")
        .replace("।", " ")
        # .replace("(", " ")
        # .replace(")", " ")
        .replace("—", " ")
        .replace("ॽ", " ")
        .replace("?", " ")
        .replace("ः", "")
        .replace(":", " ")
        .replace("“", ' ')
        .replace("”", ' ')
        .replace("‘", " ")
        .replace("’", " ")
        .replace('"', " ")
        .replace("'", " ")
        .replace("0", "०")
        .replace("1", "१")
        .replace("2", "२")
        .replace("3", "३")
        .replace("4", "४")
        .replace("5", "५")
        .replace("6", "६")
        .replace("7", "७")
        .replace("8", "८")
        .replace("9", "९")
    )

    # Remove emojis
    regrex_pattern = re.compile(
        pattern="["
        "\U0001f600-\U0001f64f"  # emoticons
        "\U0001f300-\U0001f5ff"  # symbols & pictographs
        "\U0001f680-\U0001f6ff"  # transport & map symbols
        "\U0001f1e0-\U0001f1ff"  # flags (iOS)
        "\U00002500-\U00002bef"  # chinese char
        "\U00002702-\U000027b0"
        "\U000024c2-\U0001f251"
        "\U0001f926-\U0001f937"
        "\U00010000-\U0010ffff"
        "\u2640-\u2642"
        "\u2600-\u2b55"
        "\u200d"
        "\u23cf"
        "\u23e9"
        "\u231a"
        "\ufe0f"  # dingbats
        "\u3030"
        "]+",
        flags=re.UNICODE,
    )
    text = regrex_pattern.sub("", text)

    pattern = re.compile(r'<.*?>') # HTML
    text = pattern.sub('', text)

    text = re.sub(r'([A-Za-z0-9._%+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,})', '', text) # Email

    # Remove multiple white spaces
    text = re.sub(r"\s+", " ", text)

    text = re.sub('[a-zA-Z]', '', text)

    # Text inside paranthesis
    text = re.sub(r'\([^)]*\)', '', text)

    text = " ".join(text.split())

    punctuations = list(string.punctuation)
    punctuations.remove(".")
    punctuations.remove("%")
    # punctuations.remove("(")
    # punctuations.remove(")")

    text = "".join([char for char in text if char not in punctuations])
    text = " ".join(text.split())

    return text
```

```python
perform_preprocessing("निखिल उप्रेतीको भैरव फिल्म'ले अमेरिकामा<html>sak@ghmail.comw ah.")
```

```python
def apply_cleaner(x):
    return perform_preprocessing(x)
```

```python
## Clean
df['title_cleaned'] = df['title'].apply(lambda x: apply_cleaner(x))
df['news_cleaned'] = df['news'].apply(lambda x: apply_cleaner(x))
```

```python
df['title_length'] = df['title_cleaned'].apply(lambda x: len(x.split()))
df['news_length'] = df['news_cleaned'].apply(lambda x: len(x.split()))
```

```python
df['news_length'].describe(), df['title_length'].describe()
```

```python
len(df), len(df[df['news_length'] <= 30]), len(df[df['title_length'] <= 2]), len(df[df['title_length'] > 12])
```

```python
### Remove very short news(<=30), and short and long titles(<=2 or >15)
df = df[df['news_length'] > 30]
df = df[df['title_length'] > 2]
df = df[df['title_length'] <= 12]
```

```python
print(f'Original dataset size: {original_dataset:,}')
print(f'After cleaning: {len(df):,}')
```

```python
from sklearn.model_selection import train_test_split
```

```python
train, test = train_test_split(df, test_size=0.15, shuffle=True, stratify=df['category'], random_state=22)
test, val = train_test_split(test, test_size=0.33, shuffle=True, stratify=test['category'], random_state=29)
```

```python
input_text = "/kaggle/working/train_text.txt"
model_prefix = "summarization_50000"
vocab_size = 50_000
character_coverage = 1
model_type = "bpe"
normalization_rule_name = "identity"
```

```python
spm.SentencePieceTrainer.train(
    f"--input={input_text} --model_prefix={model_prefix} --vocab_size={vocab_size} --character_coverage={character_coverage} --model_type={model_type} --normalization_rule_name={normalization_rule_name} --bos_id=2 --eos_id=3 --unk_id=1 --pad_id=0"
)
```
