```python
import pandas as pd
import sentencepiece as spm
```

```python
sp = spm.SentencePieceProcessor()
sp.load("/kaggle/input/nepali-summarization-tokenizer/summarization_50000.model")
```

```python
train = pd.read_csv('/kaggle/input/nepali-summarization-set-cleaned/summarization_set_cleaned_train.csv')
test = pd.read_csv('/kaggle/input/nepali-summarization-set-cleaned/summarization_set_cleaned_test.csv')
val = pd.read_csv('/kaggle/input/nepali-summarization-set-cleaned/summarization_set_cleaned_val.csv')
```

```python
train['title_labels_len'] = train['title'].apply(lambda x: len(sp.encode(x)))
test['title_labels_len'] = test['title'].apply(lambda x: len(sp.encode(x)))
val['title_labels_len'] = val['title'].apply(lambda x: len(sp.encode(x)))
```

```python
len(train), len(test), len(val)
```

```python
len(train[train['title_labels_len'] > 16]), len(test[test['title_labels_len'] > 15]), len(val[val['title_labels_len'] > 15]) 
```

```python
train = train[train['title_labels_len'] <= 16]
test = test[test['title_labels_len'] <= 16]
val = val[val['title_labels_len'] <= 16]
```

```python
len(train), len(test), len(val)
```

```python
train[['news', 'title']].to_csv('summarization_set_cleaned_train.csv', index=None)
test[['news', 'title']].to_csv('summarization_set_cleaned_test.csv', index=None)
val[['news', 'title']].to_csv('summarization_set_cleaned_val.csv', index=None)
```
